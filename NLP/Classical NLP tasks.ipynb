```python
import re
```

```python
# find all occurences of any character between 'a' and 'z'
re.findall("[a-z]", "$34.512 cash")
```

```python
# find all occurences of either "name" or "phone"
re.findall("(name|phone):", "My name: assassin, my phone: sorry can't tell")
```

```python
# find "lion", "lions" or "Lion" or "Lions"
re.findall("([Ll]ion)s?", "Give it to the Lions or the lion")
```

```python
# replace all lowercase letters with "x"

re.sub("[a-z]", "x", "Hey! what's up?")
```

```python
from nltk.corpus import gutenberg
from nltk.probability import FreqDist

# get raw text from "Sence and Sensibility" be Jane Austen
raw = gutenberg.raw("austen-sense.txt")
fd_letters = FreqDist(raw)

words = gutenberg.words("austen-sense.txt")
fd_words = FreqDist(words)
sas = nltk.Text(words)

# to size the freq dist plot
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 5))

# freq plot for letters from SAS
fd_letters.plot(100)
```

```python
plt.figure(figsize=(20, 5))
fd_words.plot(50)
```

```python
sas.collocations()
```

```python
# llok at long words in the text
longWords = [w for w in set(words) if len(w)>13]
longWords[:15]
```

```python
# Concordance view, also called Keywords in Context (KWIC), show the specified word with the words that surround it in text
sas.concordance("affectionately")
print()

sas.concordance('dare')
```

```python
# find similar words
sas.similar("affection")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 4))

# look at words as they appear over time in the book/document
sas.dispersion_plot(["sense", "love", "heart", "listen", "man", "woman"])
```
