###### One-to-one LSTM for Sequence Prediction (w/o TDL)

```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(len(seq), 1, 1)
y = seq.reshape(len(seq), 1)

n_neurons = length
n_batch = length
n_epoch = 1000
```

```python
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(1, 1)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
```

```python
result = model.predict(X, batch_size=n_batch, verbose=0)
for value in result:
    print(f'{value}')
```

###### Many-to-one LSTM for Sequence Prediction (w/o TDL)

Earlier we took 5 Samples, 1 time-step and 1 feature  
This time we'll take 1 sample, 5 time-steps and 1 feature

```python
X = seq.reshape(1, length, 1)
y = seq.reshape(1, length)

n_batch = 1
n_epoch = 500
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(length, 1)))
model.add(Dense(length))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
```

```python
result = model.predict(X, batch_size=n_batch, verbose=0)
for value in result[0, :]:
    print(f'{value}')
```

###### Many-to-many LSTM for Sequence Prediction (with TDL)

```python
from keras.layers import TimeDistributed
X = seq.reshape(1, length, 1)
y = seq.reshape(1, length, 1)

n_batch = 1
n_epoch = 1000
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(length, 1), return_sequences=True))
model.add(TimeDistributed(Dense(1)))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
```

```python
result = model.predict(X, batch_size=n_batch, verbose=0)
for value in result[0, :, 0]:
    print(f'{value}')
```
