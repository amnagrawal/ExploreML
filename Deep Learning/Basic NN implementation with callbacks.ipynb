```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import TensorBoard
from time import time
import numpy

numpy.random.seed(21)
```

```python
dataset = numpy.loadtxt("/home/aman/Learnings/Datasets/nlp/pima-indians-diabetes.data.csv", delimiter=",")
X = dataset[:, 0:8]
Y = dataset[:, -1]
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adadelta', metrics=['accuracy'])
tensorboard = TensorBoard(log_dir="logs/{}".format(time()), histogram_freq=0, write_graph=True)
```

```python
model.fit(X, Y, epochs=150, batch_size=10, callbacks=[tensorboard])
```

```python
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
